```python
from astropy import units as u
from astropy import constants as const
import numpy as np
import matplotlib.pyplot as plt
from astropy.modeling.models import BlackBody
```

```python
def Planck(freq,temp):
    num = 2 * const.h * (freq)**3
    den = (const.c**2) * (np.exp(const.h*(freq)/(const.k_B * temp))-1) * u.sr
    intensities = num/den

    return intensities
```

```python
temperature = 6000 * u.K
frequency = np.logspace(np.log10(1e13), np.log10(1e16), num=1000) * u.Hz
wavelength = np.linspace(1000,100000,num=1000)*u.Angstrom
#print(Planck(100000*u.Hz,10000*u.K).cgs)
astropy_planck = BlackBody(temperature=temperature)
print(np.exp(1))
```

```python
# b_v vs wavenumber
wavenumbers = np.linspace(0,12,100) /(u.micrometer)
converted_to_freq = (wavenumbers).to(u.Hz, equivalencies=u.spectral())
temperature1 = 10000 * u.K
temperature2 = 7000 * u.K
temperature3 = 3000 * u.K
#plt.figure(dpi=1200)
plt.plot(wavenumbers, Planck(converted_to_freq,temperature1).cgs,'r')
plt.plot(wavenumbers, Planck(converted_to_freq,temperature2).cgs,'b')
plt.plot(wavenumbers, Planck(converted_to_freq,temperature3).cgs,'g')
plt.legend(['T=10000K', 'T=7000K','T=3000K'],fontsize=13)
plt.title("$B_\\nu$ vs $\\tilde{\\nu}$",fontsize=20)
plt.ylabel("Intensity (erg $s^{-1}$ $sr^{-1}$ $Hz^{-1}$ $cm^{-2}$)",fontsize=13)
plt.xlabel("Wavenumber ($\mu m^{-1}$)",fontsize=13)
plt.show()
#plt.savefig('hw1_first_new.png', bbox_inches='tight')
```
